```python
from bs4 import BeautifulSoup as bs
import requests
import csv
import pandas as pd
```

```python
r=requests.get('https://www.cars24.com/buy-used-cars-bengaluru/?itm_source=Cars24Website&itm_medium=secondary_cta')
soup=bs(r.content)
print(soup.prettify())
```

```python
cars_names=soup.find_all('a')
```

```python
cars_names
```

```python
for r in cars_names:
    name=r.find(attrs={'a':'title'}).text
```
